```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

# Main Northern Outflow

```python
#Importing Data
ycut_right,xfit_right = np.genfromtxt("North(DG)_Model_Values.txt", skip_header = 1, unpack=True, usecols=[0,2])#check these
ycut_left,xfit_left = np.genfromtxt("North(DG)_Model_Values.txt", skip_header = 1, unpack=True, usecols=[0,5])

```

```python
#Defining Parabola

def parabola(x,a,b,c):
    f = a*x**2 + b*x +c
    return f
```

```python
l_opt,l_cov = curve_fit(parabola,xfit_right,ycut_right)  
p_opt,p_cov = curve_fit(parabola,xfit_left,ycut_left)
```

```python
fig = plt.figure(figsize=[2,6])
ax = fig.subplots()

ax.set_title("Quadratic Fit of Main North Cluster")
ax.set_ylabel("Intensity")
ax.set_xlabel("X Coord") #Def need to change this

ax.plot(xfit_right,parabola(xfit_right,l_opt[0],l_opt[1],l_opt[2]))
ax.scatter(xfit_right,ycut_right,s = 5, color = 'r', label = 'Right Side') #Original data points

ax.plot(xfit_left,parabola(xfit_left,p_opt[0],p_opt[1],p_opt[2]))
ax.scatter(xfit_left,ycut_left,s = 5, color = 'b', label = 'Left Side') #Original data points
```

## Parabola Fit for Both Sides

```python
#Defining Parabola

def parabola_b(x,x0,a,b,c):
    f = a*(x-x0)**2 + b*(x-x0) +c
    return f
```

```python
ycut_right,xfit_right = np.genfromtxt("North(DG)_Model_Values.txt", skip_header = 2, unpack=True, usecols=[0,2])#check these
ycut_left,xfit_left = np.genfromtxt("North(DG)_Model_Values.txt", skip_header = 2, unpack=True, usecols=[0,5])
```

```python
l_opt,l_cov = curve_fit(parabola_b,xfit_right,ycut_right)  
p_opt,p_cov = curve_fit(parabola_b,xfit_left,ycut_left)
```

```python
fig = plt.figure(figsize=[2,6])
ax = fig.subplots()

ax.set_title("Quadratic Fit of Main North Cluster")
ax.set_ylabel("Intensity")
ax.set_xlabel("X Coord") #Def need to change this

ax.plot(xfit_right,parabola_b(xfit_right,l_opt[0],l_opt[1],l_opt[2],l_opt[3]))
ax.scatter(xfit_right,ycut_right,s = 5, color = 'r', label = 'Right Side') #Original data points

ax.plot(xfit_left,parabola_b(xfit_left,p_opt[0],p_opt[1],p_opt[2],p_opt[3]))
ax.scatter(xfit_left,ycut_left,s = 5, color = 'b', label = 'Left Side') #Original data points

```
